### 차원 축소와 k평균 군집화
* iris dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# !wget https://raw.githubusercontent.com/devdio/datasets/main/iris.csv
```

```python
iris = pd.read_csv('iris.csv')
iris.shape
```

```python
df = iris.copy()
df.head()
```

```python
X = df.drop(['Id','Species'], axis = 1)
y = df['Species']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_s = scaler.fit_transform(X)
x_s
```

### 2. 차원축소

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_p = pca.fit_transform(x_s)
X_p
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
data = pd.DataFrame(X_p, columns=['PC1', 'PC2'])
data
```

```python
plt.scatter(data['PC1'], data['PC2'],c=y, s=50)
plt.show()
y
```

### K-Means 군집화

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3)
km = km.fit(x_p)
```

```python
y_pred = km.predict(X_p)
y_pred
```

```python
y
```

```python
plt.scatter(data['PC1'], data['PC2'],c=y_pred, s=50)
plt.show()
y
```

```python
plt.scatter(data['PC1'], data['PC2'],c=y, s=50)
plt.filter('iris')
plt.show()
```

### K 값 찾기

```python
inertia = []

for k in range(2, 10):
  km = KMeans(n_clusters=k)
  km.fit(X_p)
  inertia.append(km.inertia_)
```

```python
inertia
```

```python
plt.plot(np.arange(2,10), inertia, 'o')
plt.plot(np.arange(2,10), inertia, '--')
plt.show()
```
